```python
india = [1,3,5,7,9]
usa = [2,4,6,8,10]

plt.figure(figsize = (5,3))
plt.title('Scatter plot')
plt.xlabel('india_values')
plt.ylabel('usa_values')

plt.scatter(x = india, y = usa, color = 'r', marker = 'd')
plt.show()
```

```python
# linechart

india = [1,3,5,7,9]
usa = [2,4,6,8,10]

plt.figure(figsize = (5,3))
plt.title('Scatter plot')
plt.xlabel('india_values')
plt.ylabel('usa_values')

plt.plot(india,usa, color = 'r', marker = 'd')
plt.show()
```

```python
india = [1,8,5,4,9]
usa = [2,4,3,7,1]

plt.figure(figsize = (5,3))
plt.title('Line plot')
plt.xlabel('india_values')
plt.ylabel('usa_values')

plt.plot(india,usa, color = 'r', marker = 'd')
plt.show()
```

```python
india = [1,8,5,4,9]
usa = [2,4,3,7,1]
canada = [2,5,4,7,3]

plt.figure(figsize = (5,3))
plt.title('Line plot')
plt.xlabel('india_values')
plt.ylabel('usa_values')

plt.plot(india,usa,canada, color = 'r', marker = 'd')

plt.plot(india,canada, color = 'g', marker = 'd')
plt.show()
```

```python
india = [1,8,5,4,9]
usa = [2,4,3,7,1]
canada = [2,5,4,7,3]

plt.figure(figsize = (5,3))
plt.title('Line plot')
plt.xlabel('india_values')
plt.ylabel('usa_values')

plt.plot(india,usa, color = 'r', marker = 'd')

plt.plot(india,canada, color = 'g', marker = 'd')
plt.show()
```

```python
# legend
india = [1,8,5,4,9]
usa = [2,4,3,7,1]
canada = [2,5,4,7,3]

plt.figure(figsize = (5,3))
plt.title('Line plot')
plt.xlabel('india_values')
plt.ylabel('usa_values')

plt.plot(india,usa, color = 'r', marker = 'd', label = 'india-usa')

plt.plot(india,canada, color = 'g', marker = 'd',label = 'india-canada')
plt.legend()
plt.show()
```

```python
# subplots
india = [1,8,5,4,9]
usa = [2,4,3,7,1]
canada = [2,5,4,7,3]
plt.figure(figsize = (8,3))

plt.subplot(1,2,1)
plt.title('india_usa')
plt.xlabel('india')
plt.ylabel('usa')
plt.plot(india,usa,color='r',marker='*')


plt.subplot(1,2,2)
plt.title('india_canada')
plt.xlabel('india')
plt.ylabel('canada')
plt.plot(india,usa,color='g',marker='d')

plt.show()
```

```python
# histogram

age = [1,3,4,5,9,11,14,16,18,22,29,27,27,33,35]
plt.figure(figsize= (5,3))
plt.hist(x = age)
plt.show()
```

```python
# Bar chart
company = ['google','msft','meta','IBM']
revenue = [1600,800,1100,1900]
plt.figure(figsize=(5,3))
plt.bar(company,revenue,color = ['r','g','black','y'])
plt.show()
```

```python
company = ['google','msft','meta','IBM']
revenue = [1600,800,1100,1900]
plt.figure(figsize=(5,3))
plt.barh(company,revenue,color = ['r','g','black','y'])
plt.show()
```

```python
# Pie chart

company = ['google','msft','meta','IBM']
revenue = [1600,800,1100,1900]
plt.figure(figsize=(5,3))
plt.pie(revenue,labels = company)
plt.show()
```

```python

company = ['google','msft','meta','IBM']
revenue = [1600,800,1100,1900]
plt.figure(figsize=(5,3))
plt.pie(revenue,labels = company, shadow = True, autopct= '%.2f')
plt.show()
```
